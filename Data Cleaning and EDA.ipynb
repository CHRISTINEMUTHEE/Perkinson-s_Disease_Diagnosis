## Cleaning the data

```python
# Renaming the columns with unknown values
cols={'Unnamed: 0':'Age',
'Unnamed: 2':'History',
'Unnamed: 3':'Tremors',
'Unnamed: 64':'Status'}
park_df.rename(columns=cols,inplace=True)
```

```python
# Renaming columns to reasonable names
def replacement(data):
    cols=['19.','20.','21.','22.','24.','25.','27.','28.','29.','30.','31.']
    for i in cols:
        data.columns=data.columns.str.replace(i,'')
    return data.columns
```

```python
# Renaming columns with numbers in their names
replacement(park_df)
```

```python
def shape(data):
    state=print("My dataset has ",data.shape[0]," rows and ",data.shape[1],"columns")
    return state
```

```python
# Checkigng the shape of our dataset before viewing the duplicates
shape(park_df)
```

```python
# Checking for duplicates in our dataset and dropping them
def duplis(data):
    dups=data.duplicated().sum()
    if dups > 0:
        df=data.drop_duplicates()
        return df
```

```python
# Rechecking for duplicates in the dataset
df=duplis(park_df)
shape(df)
```

```python
# Checking the percentage of in each column's missing values
def percent_missing (data):
    missing=(df.isnull().sum()/len(df)*100)
    missing_df=pd.DataFrame({'Percentage_missing':missing},index=data.columns)
    return missing_df
```
