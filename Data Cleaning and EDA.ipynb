```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Checking the data
def Dataframe(data):
    df=pd.read_csv(data,header=1)
    return df
```

```python
# Loading the dataset
park_df=Dataframe('parkinson.csv')
park_df.head()
```

```python
# Checking the extreemities in the dataset
park_df.tail()
```

## Cleaning the data

```python
# Renaming the columns with unknown values
cols={'Unnamed: 0':'Age',
'Unnamed: 2':'History',
'Unnamed: 3':'Tremors',
'Unnamed: 64':'Status'}
park_df.rename(columns=cols,inplace=True)
```

```python
# Renaming columns to reasonable names
def replacement(data):
    cols=['18.','19.','20.','21.','22.','23.','24.','25.','27.','28.','29.','30.','31.']
    for i in cols:
        data.columns=data.columns.str.replace(i,'')
    data.columns=data.columns.str.replace('     ','')
    data.columns=data.columns.str.strip()
    return data.columns
```

```python
# Renaming columns with numbers in their names
replacement(park_df)
```

```python
def shape(data):
    state=print("My dataset has ",data.shape[0]," rows and ",data.shape[1],"columns")
    return state
```

```python
# Checkigng the shape of our dataset before viewing the duplicates
shape(park_df)
```

```python
# Checking for duplicates in our dataset and dropping them
def duplis(data):
    dups=data.duplicated().sum()
    if dups > 0:
        df=data.drop_duplicates()
        return df
```

```python
# Rechecking for duplicates in the dataset
df=duplis(park_df)
shape(df)
```

```python
df.head()
```

```python
# Defining our missing values
df=df.replace(('?',np.NaN))
df=df.replace(('-',np.NaN))
```

```python
# Checking the percentage of in each column's missing values
def percent_missing (data):
    missing=(df.isnull().sum()/len(df)*100)
    missing_df=pd.DataFrame({'Percentage_missing':missing},index=data.columns)
    return missing_df
```

```python
percent_missing (df)
```

```python
# checking the percentage of missing values in the dataset
df.drop(columns=['Age of disease onset (years)', 'Duration of disease from first symptoms (years)','Hoehn & Yahr scale (-)', 'UPDRS III total (-)'],axis=1,inplace=True)
```

```python
percent_missing(df).transpose()
```

```python
df.drop(df.iloc[:,10:30],axis=1,inplace=True)
```

```python
# Dropping the columns whose percentage of missing values is above 90%
percent_missing(df).transpose()
```

```python
# Dropping rows whose entries are below 80% of the data
df.dropna(thresh=37,inplace=True)
```

```python
shape(df)
```

```python
# Rechecking the % of missing values
percent_missing(df).transpose()
```

```python
df.columns
```

```python
# This function fills the
def fill_missing(data,col1,col2):
    grouped=data.groupby(col1)[col2].mean()
    if data[col1].any()=='No':
        data[col2]=data[col2].fillna(grouped[0])
    else:
        data[col2]=data[col2].fillna(grouped[1])
```

```python
fill_missing(df,'Status','Latency of\nrespiratory exchange (ms).1')
```

```python
fill_missing(df,'Status','Rate of speech respiration (- /min)')
```
